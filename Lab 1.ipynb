```python
m = 2**31 + 1
```

```python
def calculate_X(X):
    A = 5 ** 13
    return (A * X) % m
```

```python
def generate_random (N): 
    
    for n in range (N):
        if n == 0:
            X0 = 101010101
            R0 = X0 / m 
            X = calculate_X(X0)
            yield R0
            
        else:
            yield X / m
            X = calculate_X(X)
```

```python
arr = []
for i in generate_random(1000):
    arr.append(i)
```
